```python
from Bio import Entrez

# Provide your email address to the Entrez database
Entrez.email = 'your_email@address.com'

def retrieve_title_and_results(pmid):
    handle = Entrez.efetch(db='pubmed', id=pmid, retmode='xml')
    record = Entrez.read(handle)['PubmedArticle'][0]['MedlineCitation']

    # Retrieve the title and "Results" section of the abstract
    title = record['Article']['ArticleTitle']
    abstract = record['Article']['Abstract']['AbstractText']
    results = ''
    for section in abstract:
        if 'RESULTS' == section.attributes['Label']:
            results = section
            break                  
    return (title, results)

# Example usage
pmid_list = ["32909322",
             "34170046",
             "29324532",
             "31911532",
             "33201377",
             "33259994",
             "29288922",
             "29410142",
             "32202684",
             "36181563",
             "33984726",
             "32276792",
             "29808509",
             "29432903",
             "29995749",
             "31765050",
             "32798244",
             "32246761",
             "31614036",
             "31853714",
             "33177150",
             "32530118",
             "31882242",
             "29580916",
             "31023119"]
k=1
for pmid in pmid_list:
    title, results = retrieve_title_and_results(pmid)
    print(f'\n\n***{k}***\n')
    print("Please write me an abstract for a medical journal with 'BACKGROUND', 'OBJECTIVES', 'METHODS', 'RESULT' and 'CONCLUSION' sections. Base your response on the provided 'Title' and 'Results'")
    print('Title:', title)
    print('Results:', results)
    print('\n'*3)
    print(f'{pmid}.pdf\n\n')
    k+=1
    input("Press any key...")

```
